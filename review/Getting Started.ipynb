# Python + Anaconda

Anaconda is a self-contained python installation that comes pre-packaged with lots of powerful packages and tools.

1. Download and install Anaconda3 from https://www.continuium.io/ to `C:\anaconda`
2. Python should be added to the path automatically. Restart all command prompts to reload `PATH`.
3. In the command prompt, run `pip3 install --upgrade jupyter pandas sklearn networkx pymysql`

# SQL

## SQLite3

1. Download and install from https://www.sqlite.org/download.html


## MySQL

1. Download and install from http://dev.mysql.com/downloads/installer/
2. Add `C:\Program Files\MySQL\MySQL Server 5.7\bin` to `PATH`. On Windows 10:
    1. System Settings -> About -> System Info -> Advanced System Settings -> Environment Variables
    2. Click PATH and then Edit
    3. Click New and add `C:\Program Files\MySQL\MySQL Server 5.7\bin`
3. In command prompt, check proper installation by running `mysql -u username -p`
4. Install MySQL Workbench from https://www.mysql.de/products/workbench/

# Test

- Create a file within the project called `secrets.json`. From the command line, use `touch secrets.json`
- Fill the file with the connection parameters as follows: 
```json 
{
	"test_db": {
		"host": "localhost",
		"db": "mysql",
		"user": "YOUR USERNAME",
		"password": "YOUR PASSWORD",
		"port": 3306
	}
}
```
- Run `python3 test.py`

# Extras

This course most likely won't be able to use a full stack of technologies, but here's the instructions for installing them:

## Java

1. Download and install the latest version of Java 1.8 JDK from http://www.oracle.com/technetwork/java/javase/downloads/

## Eclipse + PyDev

### Install Eclipse

1. Download and install the Eclipse IDE from https://www.eclipse.org/downloads/. The "Eclipse IDE for Java Developers" option is generally the best to start with. 
2. On first start, choose `C:\workspace` as your workspace. Click the radio box to use this as the default workspace.


### Install PyDev Extension

This extension facilitates using eclipse with the python language. 

1. Follow the instructions at http://www.pydev.org/manual_101_install.html to download and install from within the Eclipse software manager. Eclipse might ask questions to install unsigned content, this is okay.
2. Restart eclipse.
3. Later, we will automatically configure which python interpreter PyDev will use. You can optionally set it up yourself here if your installation isn't standard.
    - Preferences -> PyDev -> Interpreters -> Python Interpreter
    - in terminal type `which python3` and copy paste the output into "Interpreter Executable"
    - select for the right python interpreter. for mac users, don't use system python. bad idea.

### Install EGit Extension

Eclipse can play nice with git and GitHub using the EGit extension

1. Follow the instructions from http://www.vogella.com/tutorials/EclipseGit/article.html#eclipseinstallationgit to download and install EGit from wihin the Eclipse software manager

### Importing Git Project

1. File -> Import -> Git -> Projects from Git -> Existing Local Repository
2. Click add and navigate to `C:\git\bug-free-eureka`
3. Click the box and finish
4. Select "Import as General Project"

Resources:
- http://stackoverflow.com/questions/6760115/importing-a-github-project-into-eclipse
